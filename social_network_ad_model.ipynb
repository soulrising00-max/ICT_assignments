```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=df.drop(columns=['User ID'])
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
display(df.head())
```

```python
df.corr().loc["Purchased"]
```

```python
X = df.drop('Purchased', axis=1)
y = df['Purchased']

print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

scaler = StandardScaler()

# Columns to scale
columns_to_scale = ['Age', 'EstimatedSalary']

# Apply scaler to X_train
X_train_scaled_numeric = scaler.fit_transform(X_train[columns_to_scale])
X_train_scaled_df = pd.DataFrame(X_train_scaled_numeric, columns=columns_to_scale, index=X_train.index)
X_train_scaled = pd.concat([X_train[['Gender']], X_train_scaled_df], axis=1)

# Apply scaler to X_test
X_test_scaled_numeric = scaler.transform(X_test[columns_to_scale])
X_test_scaled_df = pd.DataFrame(X_test_scaled_numeric, columns=columns_to_scale, index=X_test.index)
X_test_scaled = pd.concat([X_test[['Gender']], X_test_scaled_df], axis=1)

print("Shape of X_train_scaled:", X_train_scaled.shape)
print("Shape of X_test_scaled:", X_test_scaled.shape)

print("\nX_train_scaled (first 5 rows):")
display(X_train_scaled.head())

print("\nX_test_scaled (first 5 rows):")
display(X_test_scaled.head())
```

```python
from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model
model = LogisticRegression(random_state=42)

# Fit the model to the training data
model.fit(X_train_scaled, y_train)

print("Logistic Regression model fitted successfully.")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Make predictions on the scaled test data
y_pred = model.predict(X_test_scaled)

print("Predictions on X_test_scaled (first 10):\n", y_pred[:10])

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Store metrics in a dictionary
Logistic_model_metrics = {
    'Accuracy': accuracy,
    'Precision': precision,
    'Recall': recall,
    'F1-Score': f1
}

print("\nModel Evaluation Metrics:")
for metric, value in Logistic_model_metrics.items():
    print(f"{metric}: {value:.4f}")
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Instantiate the KNN classifier
knn_model = KNeighborsClassifier(n_neighbors=7) # You can adjust n_neighbors

# Fit the model to the training data
knn_model.fit(X_train_scaled, y_train)

print("KNN Classifier model fitted successfully.")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
#3-0.8723,5- 0.8936,7-0.9149
# Make predictions on the scaled test data using the KNN model
y_pred_knn = knn_model.predict(X_test_scaled)

print("Predictions on X_test_scaled (first 10) from KNN model:\n", y_pred_knn[:10])

# Calculate evaluation metrics for KNN model
accuracy_knn = accuracy_score(y_test, y_pred_knn)
precision_knn = precision_score(y_test, y_pred_knn)
recall_knn = recall_score(y_test, y_pred_knn)
f1_knn = f1_score(y_test, y_pred_knn)

# Store metrics in a dictionary
KNN1_model_metrics = {
    'Accuracy': accuracy_knn,
    'Precision': precision_knn,
    'Recall': recall_knn,
    'F1-Score': f1_knn
}

print("\nKNN Model Evaluation Metrics:")
for metric, value in KNN1_model_metrics.items():
    print(f"{metric}: {value:.4f}")
```

```python
from sklearn.svm import SVC

# Instantiate the SVM classifier
svm_model = SVC(kernel='rbf', random_state=42) # You can adjust the kernel and other parameters

# Fit the model to the training data
svm_model.fit(X_train_scaled, y_train)

print("SVM Classifier model fitted successfully.")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
#Recall: ploy- 0.7234 ,rbf- 0.9362
# Make predictions on the scaled test data using the SVM model
y_pred_svm = svm_model.predict(X_test_scaled)

print("Predictions on X_test_scaled (first 10) from SVM model:\n", y_pred_svm[:10])

# Calculate evaluation metrics for SVM model
accuracy_svm = accuracy_score(y_test, y_pred_svm)
precision_svm = precision_score(y_test, y_pred_svm)
recall_svm = recall_score(y_test, y_pred_svm)
f1_svm = f1_score(y_test, y_pred_svm)

# Store metrics in a dictionary
SVM_model_metrics = {
    'Accuracy': accuracy_svm,
    'Precision': precision_svm,
    'Recall': recall_svm,
    'F1-Score': f1_svm
}

print("\nSVM Model Evaluation Metrics:")
for metric, value in SVM_model_metrics.items():
    print(f"{metric}: {value:.4f}")
```

```python
import pandas as pd

# Prepare data for the DataFrame
model_summary = [
    {'Model': 'Logistic Regression', 'Test Accuracy': Logistic_model_metrics['Accuracy'], 'Test Precision': Logistic_model_metrics['Precision'], 'Test Recall': Logistic_model_metrics['Recall'], 'Test F1-Score': Logistic_model_metrics['F1-Score']},
    {'Model': 'KNN', 'Test Accuracy': KNN1_model_metrics['Accuracy'], 'Test Precision': KNN1_model_metrics['Precision'], 'Test Recall': KNN1_model_metrics['Recall'], 'Test F1-Score': KNN1_model_metrics['F1-Score']},
    {'Model': 'SVM', 'Test Accuracy': SVM_model_metrics['Accuracy'], 'Test Precision': SVM_model_metrics['Precision'], 'Test Recall': SVM_model_metrics['Recall'], 'Test F1-Score': SVM_model_metrics['F1-Score']}
]

# Create the DataFrame
summary_df = pd.DataFrame(model_summary)

display(summary_df)
```
