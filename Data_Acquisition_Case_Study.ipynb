## Q.16 In which season (Spring, Summer, Fall, Winter) people rented bikes the most?

```python
import pandas as pd

df = pd.read_csv("P:/Downloads/bike+sharing+dataset/hour.csv")
season_mapping = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
season_counts = df.groupby('season')['cnt'].sum().reset_index()
season_counts['season_name'] = season_counts['season'].map(season_mapping)
max_rentals_row = season_counts.loc[season_counts['cnt'].idxmax()]

print("Total rentals per season:")
print(season_counts[['season_name', 'cnt']])
print(f"\nThe season with the most bike rentals is: {max_rentals_row['season_name']} with {max_rentals_row['cnt']:,} rentals.")
```

## Q.17 What is the peak hour in which bike rents the most?

```python
import pandas as pd

df = pd.read_csv("P:/Downloads/bike+sharing+dataset/hour.csv")
hour_counts = df.groupby('hr')['cnt'].sum().reset_index()
peak_hour_row = hour_counts.loc[hour_counts['cnt'].idxmax()]
peak_hour = int(peak_hour_row['hr'])
total_rentals = int(peak_hour_row['cnt'])

print(f"The peak hour for bike rentals is: {peak_hour}:00 with {total_rentals:,} rentals.")
```

## Q.18 In which day of a week bikes rents out most?

```python
# Write your code here
import pandas as pd

weekday_mapping = {
    0: 'Sunday', 1: 'Monday', 2: 'Tuesday', 
    3: 'Wednesday', 4: 'Thursday', 5: 'Friday', 6: 'Saturday'
}
weekday_counts = df.groupby('weekday')['cnt'].sum().reset_index()
weekday_counts['day_name'] = weekday_counts['weekday'].map(weekday_mapping)
max_day_row = weekday_counts.loc[weekday_counts['cnt'].idxmax()]

print("Total rentals per day of the week:")
print(weekday_counts[['day_name', 'cnt']])
print(f"\nThe day with the most bike rentals is: {max_day_row['day_name']} with {max_day_row['cnt']:,} rentals.")
```

## Q.19 In which hour Casual users rents bikes the most?

```python
# Write your code here
import pandas as pd

casual_hour_counts = df.groupby('hr')['casual'].sum().reset_index()
peak_casual_hour_row = casual_hour_counts.loc[casual_hour_counts['casual'].idxmax()]
peak_hour = int(peak_casual_hour_row['hr'])
total_casual_rentals = int(peak_casual_hour_row['casual'])

print(f"The peak hour for casual bike rentals is: {peak_hour}:00 with {total_casual_rentals:,} rentals.")
```

## Q.20 What is the maximum temperature observed in each of the seasons?

```python
# Write your code here
import pandas as pd

season_mapping = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
max_temp_per_season = df.groupby('season')['temp'].max().reset_index()
max_temp_per_season['season_name'] = max_temp_per_season['season'].map(season_mapping)

print("Maximum (normalized) temperature observed per season:")
print(max_temp_per_season[['season_name', 'temp']])
```

5 questions

What is the correlation between temperature and total bike rentals?

```python
import pandas as pd

correlation = df['temp'].corr(df['cnt'])
print(f"1. Correlation between Temp and Count: {correlation:.4f}")
```

How does the weather situation affect the average number of bike rentals?

```python
weather_mapping = {1: 'Clear/Cloudy', 2: 'Mist', 3: 'Light Precip', 4: 'Heavy Precip'}
weather_avg = df.groupby('weathersit')['cnt'].mean().reset_index()
weather_avg['weather_desc'] = weather_avg['weathersit'].map(weather_mapping)
print("\n2. Average Rentals by Weather Situation:")
print(weather_avg[['weather_desc', 'cnt']])
```

What was the percentage growth in total bike rentals from 2011 to 2012?

```python
yearly_counts = df.groupby('yr')['cnt'].sum()
growth_pct = ((yearly_counts[1] - yearly_counts[0]) / yearly_counts[0]) * 100
print(f"\n3. Percentage Growth (2011 to 2012): {growth_pct:.2f}%")
```

What is the overall percentage split between Casual and Registered users?

```python
total_casual = df['casual'].sum()
total_registered = df['registered'].sum()
total_rentals = total_casual + total_registered

casual_pct = (total_casual / total_rentals) * 100
registered_pct = (total_registered / total_rentals) * 100

print(f"\n4. User Type Split: Casual: {casual_pct:.2f}%, Registered: {registered_pct:.2f}%")
```

Is the average number of bike rentals higher on working days or non-working days?

```python
workday_avg = df.groupby('workingday')['cnt'].mean().reset_index()
workday_labels = {0: 'Non-Working Day', 1: 'Working Day'}
workday_avg['day_type'] = workday_avg['workingday'].map(workday_labels)
print("\n5. Average Rentals by Day Type:")
print(workday_avg[['day_type', 'cnt']])
```
